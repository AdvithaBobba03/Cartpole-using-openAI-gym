```python
import gym

# Create the CartPole environment with render_mode specified
env = gym.make('CartPole-v1', render_mode="human")
env.reset()

# Loop through episodes
for episode in range(1, 11):
    score = 0
    state = env.reset()
    done = False
    
    while not done:
        env.render()  # Render the environment with the specified mode
        action = env.action_space.sample()
        step_result = env.step(action)  # Store the step result
        next_state, reward, done, info = step_result[:4]  # Unpack only the first four values
        score += reward
        
    print('Episode:', episode, 'Score:', score)

env.close()
```

Model training

```python
env = gym.make(env_name)
env = DummyVecEnv([lambda: env])
model = PPO('MlpPolicy', env, verbose=1)
```

```python
model.learn(total_timesteps=20000)
```

```python
# save the model
model.save('ppo model')
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```
